# Obtenção dos dados

Usaremos o dataset da Our World in Data, que são atualizados pela Johns Hopkins University (JHU). Estes dados são coletados de fontes oficiais e compilados pelo Our World in Data.

O dataset do Our World in Data é uma fonte abrangente de dados sobre a COVID-19, contendo informações diárias sobre casos confirmados, mortes e testes em todo o mundo. Além disso, o conjunto de dados também inclui informações sobre vacinação, políticas de restrição de mobilidade e outras variáveis ​​relevantes para a pandemia.

Além disso, há conflitos existentes entre dados do governo federal e estaduais, bem como entre dados oficiais e não oficiais. Esses conflitos podem levar a discrepâncias e erros nos dados e tornar a análise mais difícil.

# Diconário de variáveis

**iso_code**: trata-se de seu código de 3 letras, definido pela norma ISO 3166-1 alpha-3. Quando é um agrupamento também é um código de 3 letras prefixado por "OWID_". Também são prefixados por OWID_ os campos correspondentes aos 4 países constituintes do Reino Unido (Inglaterra, Escócia, País de Gales e Irlanda do Norte). Exemplos de valores para esta coluna: (BRA: Brasil, DEU: Alemanha, OWID_AFR: África, OWID_WRL: Mundo, OWID_HIC: Países de alta renda.  
**continent**:  continente no qual o país se localiza. Quando se trata de agrupamento, esta coluna vem em branco.  
**location**: nome do país ou agrupamento (em inglês)  
**date**: data do registro  
**total_cases**: total de casos   
**new_cases**: novos casos no dia   
**new_cases_smoothed**: média móvel de 7 dias de casos  
**total_deaths**: total de mortes   
**new_deaths**: novas mortes  
**new_deaths_smoothed**: média móvel de 7 dias de mortes  
**total_cases_per_million**: total de casos por milhão de pessoas  
**new_cases_per_million**:  Novos casos por milhão  
**new_cases_smoothed_per_million**: média móvel de casos por milhão  
**total_deaths_per_million**: total de mortes por milhão  
**new_deaths_per_million**: novas mortes por milhão  
**new_deaths_smoothed_per_million**: média móvel de novas mortes por milhão  
**reproduction_rate**: estimativa em tempo real da taxa de reprodução da COVID-19 (veja este [link](https://github.com/crondonm/TrackingR/tree/main/Estimates-Database)).  
**icu_patients**: internações em UTI  
**icu_patients_per_million**: pacientes internados em UTI por milhão  
**hosp_patients**: internações hospitalares  
**hosp_patients_per_million**: internações hospitalares por milhão  
**weekly_icu_admissions**: número de novos pacientes admitidos em uma semana (na data corrente e nos 6 dias anteriores)  
**weekly_icu_admissions_per_million**: número de novas admissões em UTI em uma semana por milhão pessoas  
**weekly_hosp_admissions**: número de novas internações hospitalares no período de uma semana  
**weekly_hosp_admissions_per_million**: número de novas internações hospitalares no período de uma semana  
**total_tests**: total de testes de COVID-19 realizados  
**new_tests**: número de novos testes no dia (calculado somente para dias consecutivos)  
**total_tests_per_thousand**: total de testes por mil pessoas  
**new_tests_per_thousand**: novos testes por 1000 pessoas  
**new_tests_smoothed**: média móvel de 7 dias de testes realizados  
**new_tests_smoothed_per_thousand**: média móvel de 7 dias de testes realizados por 1009 pessoas  
**positive_rate**: proporção de testes que dão resultado positivo, em uma média móvel de 7 dias. É o inverso da coluna abaixo  
**tests_per_case**: total de testes por caso positivo em uma média móvel de 7 dias. É o inverso da colunan **positive_rate** acima  
**tests_units**: coluna categórica que indica a metodologia de teste utilizada por cada país Pode ter os seguintes valores: *people tested* (pessoas testadas), *tests performed* (testes efetuados, uma pessoa pode ser testada mais de uma vez) ou *samples tested* (amostras testadas, em alguns casos mais de uma amostra pode ser necessária para efetuar um teste)    
**total_vaccinations**: Total de vacinas aplicadas  
**people_vaccinated**: Total de pessoas vacinadas  
**people_fully_vaccinated**: Total de pessoas que tomaram todas as doses previstas pelo protocolo de vacinação de cada país   
**total_boosters**: Total de doses de reforço (doses aplicadas além do previsto pelo protocolo de vacinação de cada país)  
**new_vaccinations**: novas vacinas aplicadas no dia  
**new_vaccinations_smoothed**: média móvel de 7 dias de novas vacinas  
**total_vaccinations_per_hundred**: total de vacinas aplicadas por 100 pessoas  
**people_vaccinated_per_hundred**: total de indivíduos vacinados por 100 pessoas  
**people_fully_vaccinated_per_hundred**: total de pessoas totalmente vacinadas por 100  
**total_boosters_per_hundred**: doses de reforço por 100 pessoas  
**new_vaccinations_smoothed_per_million**: média móvel de 7 dias de novas vacinas aplicadas por milhão  
**new_people_vaccinated_smoothed**: média móvel de 7 dias de novos indivíduos vacinados  
**new_people_vaccinated_smoothed_per_hundred**: média móvel de 7 dias de novos indivíduos vacinados por 100  
**stringency_index**: Índice, composto de 9 variáveis que denota o grau de restrição imposto pelos governos frente à pandemia. É um valor que varia de 0 a 100 (restrições mais rígidas)  
**population_density**: densidade populacional do local, ou seja, total de pessoas divididas pela área   
**median_age**: mediana da idade da população usando uma projeção da ONU de 2020  
**aged_65_older**: proporção de indivíduos com 65 anos ou mais no ano mais recente disponível   
**aged_70_older**: proporção de indivíduos com 70 anos ou mais no ano de 2015  
**gdp_per_capita**: PIB per capita utilizando o método da [Paridade do Poder de Compra](https://pt.wikipedia.org/wiki/Paridade_do_poder_de_compra) (PPC)   
**extreme_poverty**: proporção de pessoas vivendo em condição de extrema pobreza  
**cardiovasc_death_rate**: taxa de morte por doenças cardiovasculares em 2017 (total de mortes por 100000 pessoas)  
**diabetes_prevalence**: prevalência de diabetes em porcentagem de pessoas entre 20 e 79 anos  
**female_smokers**: proporção de mulheres fumantes na população  
**male_smokers**:  proporção de homens fumantes na população  
**handwashing_facilities**: proporção de pessoas com acesso a instalações adequadas para lavagem de mãos  
**hospital_beds_per_thousand**: letos de hospital por 1000 habitantes  
**life_expectancy**: expectativa de vida ao nascer em 2019  
**human_development_index**: Índice de Desenvolvimento Humano (IDH). Índice estatístico composto por três variáveis: renda, escolaridade e expectativa de vida.   
**population**: população total da localidade   
**excess_mortality_cumulative_absolute**: diferença porcentual entre o número de mortes desde 1º de janeiro de 2020 e a projeção do total de mortes para o mesmo período em anos anteriores   
**excess_mortality_cumulative**: diferença porcentual entre o acumulado de mortes desde 1º de janeiro de 2020 e a projeção do acumulado de mortes para o mesmo período em anos anteriores  
**excess_mortality**: diferença porcentual entre mortes semanais ou mensais reportadas em 2020-21 e o número projetado de mortes para o mesmo período em anos anteriores  
**excess_mortality_cumulative_per_million**: diferença porcentual entre o número de mortes desde 1º de janeiro de 2020 e a projeção do total de mortes para o mesmo período em anos anteriores por milhão de pessoas

Vamos plotar um grafico mostrando os 5 paises com mais mortes

```python
# Vamos colocar a data em uma variavel
df_date = "2023-03-06"

# Separar separar os paises dos agrupamentos. 
df_paises = df[(df.iso_code.str.contains("OWID") == False) & (df.date <= df_date)]


# gráfico de barras para total de mortes 
top_5_total_deaths = df_paises.loc[df.date == '2023-03-06', ['location', 'total_deaths']].sort_values(by="total_deaths", ascending=False)[0:5]

# Plotar na escala em milhões
top_5_total_deaths['total_deaths_milhoes'] = top_5_total_deaths['total_deaths'] / 1000000

fig, ax = plt.subplots(figsize=(12,8))
ax.barh('location', 'total_deaths_milhoes', data=top_5_total_deaths)
ax.set_title("Gráficos de total de mortes")
ax.set_xlabel("N° mortes (em milhões)") 



plt.tight_layout()
```

Os países com mais mortes são países superpopulosos. Estados Unidos lidera esta marca triste com mais de 1.1 milhão de mortes.

```python
# gráfico de barras para total de casos
top_5_total_cases = df_paises.loc[df.date == '2023-03-06', ['location', 'total_cases']].sort_values(by="total_cases", ascending=False)[0:5]

# Plotar na escala em milhões
top_5_total_cases['total_cases_milhoes'] = top_5_total_cases['total_cases'] / 1000000

fig, ax = plt.subplots(figsize=(12,8))
ax.barh('location', 'total_cases_milhoes', data=top_5_total_cases)
ax.set_title("Gráficos de total de casos")
ax.set_xlabel("N° casos (em milhões)") 



plt.tight_layout()

```

No número de casos além de países superpopulosos temos a França e a Alemanha, porém estes países não aparecem no top 5 em mortes, isto se dá, pois, estes países conseguiram diminuir sua taxa de mortalidade.


Agora vamos plotar um gráfico para ver a evolução dos casos ao longo do tempo.


```python

# Filtra os dados para o mundo inteiro
df = df[df['location'] == 'World']

# Cria o gráfico de linha
fig = px.line(df, x='date', y='total_cases', title='Evolução dos casos de COVID-19 no mundo')
fig.update_yaxes(title='casos (em milhões)')

fig.show()
```

Diante deste gráfico que mostra a evolução de casos do Covid podemos perceber algumas coisas:
* Até janeiro de 22 tivemos um aumento continuo nos casos, como sabíamos muito pouco sobre a doença, tomamos diversas medidas para reduzir a velocidade de contaminação, porém analisando o gráfico vemos que a infeção se manteve uniforme.
* Em jan. 22 tivemos o começo da segunda onda, o aumento de casos foi considerável, neste momento muitos já estavam vacinados, isto pode explicar o aumento dos casos, as pessoas ficaram mais confiantes para se expor ao risco da contaminação acreditando na imunização da vacina.


Em seguida, vamos plotar dois graficos de linha, mostrando a evolução de mortes e vacinas no mundo desde o começo da pandemia

```python
from plotly.subplots import make_subplots
# Filtra os dados para o mundo inteiro
df = df[df['location'] == 'World']

# Cria o primeiro gráfico de linha
fig1 = px.line(df, x='date', y='total_deaths', title='Evolução de mortes de COVID-19 no mundo')
fig1.update_yaxes(title='casos (em milhões)')

# Cria o segundo gráfico de linha
fig2 = px.line(df, x='date', y='people_vaccinated', title='Evolução das vacinas de COVID-19 no mundo')
fig2.update_yaxes(title='vacinas (em milhões)')

# Configura a figura para plotar os dois gráficos lado a lado
fig = make_subplots(rows=1, cols=2, subplot_titles=("Evolução de mortes de COVID-19", "Evolução das vacinas de COVID-19"))
fig.add_trace(fig1.data[0], row=1, col=1)
fig.add_trace(fig2.data[0], row=1, col=2)

# Configura os eixos e a escala
fig.update_xaxes(title_text="Data", row=1, col=1)
fig.update_xaxes(title_text="Data", row=1, col=2)
fig.update_yaxes(title_text="casos (em milhões)", row=1, col=1)
fig.update_yaxes(title_text="vacinas (em milhões)", row=1, col=2)

# Mostra o gráfico
fig.show()
```

Analisando estes gráficos percebemos que:
* A evolução das mortes foi uniforme até o 1° semestre de 2022, a curva só teve um declínio quando a vacinação atingiu seu pico.
* Pelo gráfico de evolução das vacinas, percebemos que a inclinação da curva é maior que a de mortes, isto se dá por conta do sendo de urgência, o mundo todo estava desenvolvendo vacinas. Demoramos praticamente 1 ano para conseguir vacinar a primeira pessoa. Se pegarmos o histórico de vacinas desenvolvidas, a da covid foi desenvolvida em tempo recorde.

Adiante teremos uma demonstração de como a vacina afetou a taxa de mortalidade.

Agora vamos identificar quando foi o primeiro caso no Brasil, a data da primeira morte e a diferença de dias entre o primeiro caso e a primeira morte.


```python
# Localizando o primeiro caso 
primeiro_caso = df_brasil.loc[df_brasil['new_cases'] != 0.0].head(1) 
primeiro_caso
```

```python
# Localizando a primeira morte
primeira_morte = df_brasil.loc[df_brasil['new_deaths'] != 0.0].head(1) 
primeira_morte
```

```python
# identificaçãso do primeiro caso
from datetime import datetime
data_str = primeiro_caso.date.loc[34891]
data_obj = datetime.strptime(data_str, '%Y-%m-%d')
data_formatada = data_obj.strftime('%d/%m/%Y')
print("O primeiro caso no Brasil foi em: {}".format(data_formatada))

# identificação da primeira morte
data_str2 = primeira_morte.date.loc[34912]
data_obj2 = datetime.strptime(data_str2, '%Y-%m-%d')
data_formatada2 = data_obj2.strftime('%d/%m/%Y')
print("A primeira morte no Brasil foi em: {}".format(data_formatada2))

# convertendo as strings de data em objetos de data
df_brasil['date'] = pd.to_datetime(df_brasil['date'])
# calculando a diferença em dias entre as datas
delta = df_brasil.date.loc[34888] - df_brasil.date.loc[34867]
print('Se passaram {} dias entre o primeiro caso e a primeira morte no Brasil.'.format(delta.days))


```

Para o Brasil, o primeiro caso foi em 27/02/2020. Apenas 21 dias depois tivemos a primeira morte, isto só reforça a velocidade de contaminação da doença que pode ser transmitida pelo ar, por isso, tivemos uma contaminação em massa mesmo com muitos esforços para frear a doença.

Abaixo teremos 3 gráficos mostrando a evolução de casos, mortes e vacinações.

```python
from plotly.subplots import make_subplots


# Filtra os dados para o Brasil
df_brasil = df[df['location'] == 'Brazil']

# Cria o primeiro gráfico de linha
fig1 = px.line(df_brasil, x='date', y='total_deaths', title='Evolução de mortes de COVID-19 no Brasil')
fig1.update_yaxes(title='casos (em milhões)')

# Cria o segundo gráfico de linha
fig2 = px.line(df_brasil, x='date', y='total_cases', title='Evolução dos casos de COVID-19 no Brasil')
fig2.update_yaxes(title='vacinas (em milhões)')

# Cria o terceiro gráfico de linha
fig3 = px.line(df_brasil, x='date', y='people_vaccinated', title='Evolução das vacinas contra COVID-19 no Brasil')
fig3.update_yaxes(title='vacinas (em milhões)')

# Configura a figura para plotar os três gráficos em uma coluna
fig = make_subplots(rows=1, cols=3, subplot_titles=("Evolução de mortes de COVID-19", "Evolução dos casos de COVID-19", "Evolução das vacinas contra COVID-19"))
fig.add_trace(fig1.data[0], row=1, col=1)
fig.add_trace(fig2.data[0], row=1, col=2)
fig.add_trace(fig3.data[0], row=1, col=3)

# Configura os eixos e a escala para cada gráfico
fig.update_xaxes(title_text="Data", row=1, col=1)
fig.update_yaxes(title_text="casos (em milhões)", row=1, col=1)
fig.update_xaxes(title_text="Data", row=2, col=1)
fig.update_yaxes(title_text="vacinas (em milhões)", row=2, col=1)
fig.update_xaxes(title_text="Data", row=3, col=1)
fig.update_yaxes(title_text="vacinas (em milhões)", row=3, col=1)

# Mostra o gráfico completo
fig.show()
```

Como para o mundo, as curvas são muito parecidas, da mesma forma como a vacinação foi mais rápida que a evolução de mortes e casos.

Abaixo veremos como a vacina afetou a taxa de mortalidade.


```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go

# Filtrar os dados para o Brasil
df_brasil = df[df['location'] == 'Brazil'].copy()

# Calcular a taxa de mortalidade e a porcentagem da população vacinada
df_brasil.loc[:, 'mortality_rate'] = (df_brasil['total_deaths'] / df_brasil['total_cases']) * 100
df_brasil.loc[:, 'vaccination_percentage'] = (df_brasil['people_vaccinated'] / df_brasil['population']) * 100

# Criar um gráfico com dois eixos y
fig = make_subplots(specs=[[{"secondary_y": True}]])

# Adicionar o gráfico de linha para a taxa de mortalidade
fig.add_trace(go.Scatter(x=df_brasil['date'], y=df_brasil['mortality_rate'], name='Taxa de mortalidade'), secondary_y=False)

# Adicionar o gráfico de linha para a porcentagem da população vacinada
fig.add_trace(go.Scatter(x=df_brasil['date'], y=df_brasil['vaccination_percentage'], name='Porcentagem da população vacinada'), secondary_y=True)

# Configurar os títulos e eixos dos gráficos
fig.update_layout(title='Relação entre pessoas vacinadas e taxa de mortalidade de COVID-19 no Brasil', xaxis_title='Data', yaxis_title='Taxa de mortalidade', yaxis2_title='Porcentagem da população vacinada')

fig.show()
```

Veremos como a vacina afetou a taxa de mortalidade:
* Em maio de 2020, antes do lockdown temos uma taxa de mortalidade de aproximadamente 7%, com o isolamento conseguimos diminuir esta taxa.
* Vemos claramente como a taxa de mortalidade e a porcentagem da população vacinada são inversamente proporcionais, nos mostrando a eficácia da vacina.
* Com mais de 80% da população imunizada a taxa de mortalidade chegou a 2%.
